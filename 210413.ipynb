"friendship" data: list of paris of IDs (friend relationships)

```python
friendships = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
              (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

when we want to add a list of friends to each other,
    1. We set each user's friends property to an empty list:

```python
for user in users :
    user['friends'] = []
```

    2. populate the lists using the friendships data

```python
for i, j in friendships :
    # this works because users[i] is the user whose id is i
    users[i]['friends'].append(users[j])
    users[j]['friends'].append(users[i])
```

### ? What's the average number of connections?

return total_number_of_connections with function

```python
def number_of_friends (user) :
    """ how many friends does user have?"""
    return len (user['friends'])
```

```python
total_connections = sum (number_of_friends(user)
                        for user in users)
```

```python
total_connections
```

divide by the #of users

```python
from __future__ import division
num_users = len(users)
avg_connections = total_connections / num_users
```

### ? who has the most connections? who has the least connections? 

create a list (user_id, number of friends)

```python
num_friends_by_id = [(user['id'], number_of_friends (user))
                    for user in users]
```

```python
num_friends_by_id
```

sort list (user_id, number of friends) in order to number of friends

```python
sorted(num_friends_by_id, 
       key = lambda num_friends_by_id:num_friends_by_id[1], 
       reverse = True)
```
